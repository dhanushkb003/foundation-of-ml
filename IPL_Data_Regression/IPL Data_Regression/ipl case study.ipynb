```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.boxplot(df['SOLD_PRICE'])
plt.title('Boxplot of SOLD_PRICE')
plt.ylabel('Sold Price')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['SOLD_PRICE'], bins=15)
plt.title('Distribution of SOLD_PRICE')
plt.xlabel('Sold Price')
plt.ylabel('Frequency')
plt.show()
```

```python
columns = ['SOLD_PRICE', 'BASE_PRICE', 'ODI_RUNS_S', 'T_RUNS', 'WKTS']# only for important expected variables

for col in columns:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
columns = ['SOLD_PRICE', 'BASE_PRICE', 'ODI_RUNS_S', 'T_RUNS', 'WKTS']

for col in columns:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=15)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for col in ['SOLD_PRICE', 'BASE_PRICE', 'ODI_RUNS_S', 'T_RUNS', 'WKTS']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")

# we are not going to remove any outliers as irl ipl usually have high value player which is not an error so we procced with the analysis
```

```python
correlation = df.corr(numeric_only=True)
correlation
```

```python
correlation['SOLD_PRICE']
```

```python
correlation['SOLD_PRICE'].drop('SOLD_PRICE').sort_values(ascending=False)# from highest to lowest and not considering sold price itself as it will always have 1
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(correlation, cmap='coolwarm',annot=True,fmt=".2f")
plt.show()
```

```python
plt.figure(figsize=(6,10))

sns.heatmap(
    correlation[['SOLD_PRICE']].sort_values(by='SOLD_PRICE', ascending=False),
    annot=True,
    cmap='coolwarm'
)

plt.show()

# specific for sold price for better understanding 
```
