```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Engg_College_data.csv")
```

```python
df.head()
```

```python
df1=df.drop(['SR_NO','Engg_College'],axis=1)
```

```python
df1.head()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
wardlink=linkage(df1,method='ward')
dend=dendrogram(wardlink)
```

```python
from scipy.cluster.hierarchy import fcluster
cluster1=fcluster(wardlink,2,criterion='maxclust')
cluster1
```

```python
df1['added_Clusters']=cluster1
```

```python
df1.head()
```

```python
df1['added_Custer']=cluster1
```

```python
df.head()
```

```python
df1.groupby('added_Custer').count()
```

```python
df1.groupby('added_Custer').mean()
```
